```python
#Fit a LASSO model with the raw traindata
lassocv = LassoCV()
lassocv.fit(traindata[list(range(122))], traindata[[122]])
#Fit a LASSO model with the normalized traindata
lassocv2 = LassoCV(normalize = True)
lassocv2.fit(traindata[list(range(122))], traindata[[122]])
select1,select2=[],[]
for i in range(122):
    if lassocv.coef_[i]!=0:
        select1+=[i+1]
    if lassocv2.coef_[i]!=0:
        select2+=[i+1]
print('test error is: ',mean_squared_error(testdata[[122]], lassocv.predict(testdata[list(range(122))])))
print('the variables selected by the 1st model is:',select1)
print('test error with normalized data is: ',mean_squared_error(testdata[[122]], lassocv2.predict(testdata[list(range(122))])))
print('the variables selected by the 2nd model is:',select2)
print('the test errors of two models does not have obvious difference.')
```

```python
#fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
pca = PCA()
regr = LinearRegression()
X_reduced = pca.fit_transform(scale(traindata[list(range(122))]))
mse = []; best_M=0;maximum=1;
# Calculate MSE with only the intercept (no principal components in regression)
score = -1*model_selection.cross_val_score(regr, np.ones((len(X_reduced),1)),traindata[[122]],cv=10,scoring='neg_mean_squared_error').mean()    
mse.append(score)
best_M=0;minimum=score;
# Calculate MSE using CV for the 19 principle components, adding one component at the time.
for i in np.arange(1, len(pca.components_)+1):
    score = -1*model_selection.cross_val_score(regr, X_reduced[:,:i],traindata[[122]],cv=10,scoring='neg_mean_squared_error').mean()
    mse.append(score)   
    if score<minimum:
        minimum=score;best_M=i;    
# Plot results    
plt.plot(mse, '-v')
plt.xlabel('Number of principal components in regression')
plt.ylabel('MSE')
plt.xlim(xmin=-1);
```

```python
print('the smallest cross-validation error occurs when M=',best_M)
regr.fit(X_reduced[:,:best_M], traindata[[122]])
# Prediction with test data
X_reduced_test = pca.fit_transform(scale(testdata[list(range(122))]))
pred = regr.predict(X_reduced_test[:,:best_M])
print('the test error is:',mean_squared_error(testdata[[122]], regr.predict(X_reduced_test[:,:best_M])))
```
